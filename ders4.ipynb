```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv("diabetes.csv")
df.copy()
```

```python
df.head()
```

```python
y=df["Outcome"]
X=df.drop(["Outcome"],axis=1)
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#yapay sinir aÄŸlarÄ± 
#model tahmin
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlpc_model=MLPClassifier().fit(X_train,y_train)
```

```python
mlpc_model.coefs_
```

```python
?mlpc_model
```

```python
y_pred=mlpc_model.predict(X_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
# model turing 
```

```python
mlpc_params={"alpha":[0.1,0.01,0.03,0.005,0.0001],"hidden_layer_sizes":[(10,10),(100,100,100),(100,100),(3,5)]}
```

```python
mlpc=MLPClassifier(solver="lbfgs",activation="logistic")
```

```python
mlpc_cv_model=GridSearchCV(mlpc,mlpc_params,cv=5,n_jobs=-1,verbose=2).fit(X_train,y_train)
```
